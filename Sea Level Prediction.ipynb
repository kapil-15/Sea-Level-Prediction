```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy import stats
```

```python
df= pd.read_csv('epa-sea-level.csv')
```

```python
plt.scatter(df.Year,df['CSIRO Adjusted Sea Level'], marker='x')
```

```python
# reg= linregress.LinearRegression()
```

```python
reg= linregress(df.Year,df['CSIRO Adjusted Sea Level'])

# regz = stats.linregress(df_new.Year,df_new['CSIRO Adjusted Sea Level'])
```

# best fit line

```python
plt.scatter(df.Year,df['CSIRO Adjusted Sea Level'], marker='x')


x = pd.Series([i for i in range(1880,2050)])
y = reg.intercept + reg.slope*x
plt.plot(x, y, 'r')
```

```python
new_df=df.loc[df['Year']>=2000]
x2=new_df['Year']
y2=new_df['CSIRO Adjusted Sea Level']
reg2= linregress(x2,y2)
x_pred = pd.Series(i for i in range(2000,2050))
y_pred = reg2.slope*x_pred+reg2.intercept
plt.plot(x_pred,y_pred,'green')
```

```python
# plt.figure(figsize=(12,12))
plt.scatter(df.Year,df['CSIRO Adjusted Sea Level'], marker='x')
plt.plot(df.Year,reg.intercept + reg.slope*df.Year
         ,label='first line of best fit', color='orange', linewidth=2)
plt.plot(x_pred,y_pred
         ,label='second line of best fit', color='r', linewidth=2)
plt.legend()
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.show()
```
